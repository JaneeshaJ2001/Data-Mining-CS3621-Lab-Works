```python
data = pd.read_csv('/kaggle/input/apriory/supermarket.csv')
columns_with_only_question_mark = [col for col in data.columns if (data[col] == '?').all()]
data = data.drop(columns=columns_with_only_question_mark)
data = data.drop(columns='total')
data = data.replace({'?': 0, 't': 1})
columns_to_keep = list(filter(lambda x: not x.startswith('department'), data.columns))
df1=data[columns_to_keep]
```

```python
# Find frequent itemsets with minimum support threshold
min_support = 0.4  # Adjust as needed
frequent_itemsets = apriori(data, min_support=min_support, use_colnames=True)

min_confidence = 0.7  # Adjust as needed
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)

# Print the association rules
#print(rules.to_string())

rules_sorted = rules.sort_values(by='confidence', ascending=False)

# Print the sorted association rules
print(rules_sorted.to_string(index=False))
```
